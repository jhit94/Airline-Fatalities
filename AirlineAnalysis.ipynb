```python
import pandas as pd
```

```python
original_df = pd.read_csv("https://raw.githubusercontent.com/fivethirtyeight/data/master/airline-safety/airline-safety.csv")
original_df
```

```python
sorted_df = original_df.sort_values(by='fatalities_00_14', ascending=False)
```

```python
sorted_df = sorted_df.reset_index(drop=True)
```

```python
top_ten_df = sorted_df.head(10)
```

```python
top_ten_df
```

```python
top_ten_df
```

```python
top_ten_df.plot(x='airline', y= "fatalities_00_14", kind = "bar")
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['font.size'] = 9.0
top_ten_df = top_ten_df.head()
labelsx = top_ten_df["airline"]
sizesy = top_ten_df["fatalities_00_14"]

#labels = ['Malaysia', 'American', 'France', 'kenya', 'China', 'TAM', 'India', 'Gulf Air', 'SAS', 'United/Continental',]
#sizes = [21, 16, 13, 11, 8, 7, 6, 5, 4, 4]

explode = (0.1, 0.1, 0, 0, 0)  # only "explode" the 2nd slice

fig1, ax1 = plt.subplots()
ax1.pie(sizesy, explode=explode, labels=labelsx, autopct='%1.1f%%', wedgeprops={'edgecolor': 'black'} , startangle=90)
ax1.axis('equal') 
plt.title('Top 5 Airlines with Fatalities (2014)')
plt.show()

```
